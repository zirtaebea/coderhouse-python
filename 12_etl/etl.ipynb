# Schedule
Cronograma (ou schedule, em inglÃªs) refere-se a um plano ou calendÃ¡rio que determina quando e com que frequÃªncia o processo ETL deve ser executado. O cronograma Ã© uma parte crÃ­tica do processo ETL, pois ajuda a garantir que os dados sejam atualizados regularmente.

O cronograma pode ser configurado para executar o processo ETL em:
- **intervalos regulares**, como diariamente, semanalmente ou mensalmente. 
- **baseado em eventos ou em pipeline**, como a disponibilidade de novos dados ou a conclusÃ£o de uma tarefa especÃ­fica em um sistema externo. 



```python
import schedule
import time

# variÃ¡vel para indicar para parar o schedule
parar_schedule = False

# funÃ§Ãµes 
# tarefa do schedule
def minha_tarefa():
    print("Executando minha tarefa")

# parar schedule
def parar_agendamento():
    global parar_schedule
    parar_schedule = True # a funÃ§Ã£o muda o false para true e quando parar schedule for true no while, o programa sairÃ¡ do loop

# criando evento para executar tarefa a cada 1 segundo
schedule.every(1).seconds.do(minha_tarefa)

# criando evento para parar tarefa depois de 5 segundos
schedule.every(5).seconds.do(parar_agendamento)

while True:
    if parar_schedule: # se parar schedule for true
        notification.notify( # notificaÃ§Ã£o de que o schedule foi finalizado
            title='Schedule finalizado',
            message='Mensagem de teste',
            app_name='Nome do app',
            timeout=15
)
        break # break do loop
    schedule.run_pending() #ativando schedule
    time.sleep(1)
```
